```python
from scipy import stats, special
import matplotlib.pyplot as plt
import numpy as np
```

Q1b:

```python
def posterior_1b(y, theta):
    if y<0 or y>100:
        print("parameter y is out of range")
        return
    else:
        return special.binom(100, y) * theta**y * (1-theta)**(100-y)
```

```python
theta = [i*0.1 for i in range(1,11)]
p_y = [posterior_1b(57, th) for th in theta]
```

```python
plt.plot(theta, p_y)
plt.xticks(theta)
plt.xlabel("theta")
plt.ylabel("posterior probability P(Y=57|theta)")
plt.show()
```

Q1c:

```python
Q1C_CONSTANT = sum([stats.binom(100,th).pmf(57) for th in theta])
```

```python
plt.plot(theta, p_y/Q1C_CONSTANT)
plt.xticks(theta)
plt.xlabel("theta")
plt.ylabel("posterior probability P(theta|Y=57)")
plt.show()
```

Q1d:

```python
theta = np.linspace(0,1,1001)
p_y = [posterior_1b(57, th) for th in theta]
```

```python
plt.plot(theta, p_y)
plt.xlabel("theta")
plt.ylabel("posterior probability P(Y=57|theta)\n with uniform prior")
plt.show()
```

Q1e:

```python
theta = np.linspace(0,1,1001)
p_y = [stats.beta.pdf(th, 57+1, 100-57+1) for th in theta]
```

```python
plt.plot(theta, p_y)
plt.xlabel("theta")
plt.ylabel("posterior probability P(theta|Y=57)\n with uniform prior")
plt.show()
```
