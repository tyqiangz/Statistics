```python
from scipy import special, stats
from math import sqrt, exp
import numpy as np
import matplotlib.pyplot as plt
```

\begin{align*}
    \theta 
        &\sim \operatorname{Beta}(0.5,0.5)\\
    p_J(\psi) 
        &=\frac{\mathrm{e}^{\frac{1}{2}\psi}\left(1+\mathrm{e}^{\psi}\right)^{-1}}{B(\frac{1}{2},\frac{1}{2})}
\end{align*}

```python
def prior_theta(th):
    return stats.beta.pdf(th, 0.5, 0.5)

def prior_psi(psi):
    return exp(0.5*psi) / (1+exp(psi)) / special.beta(0.5,0.5)
```

```python
theta = np.arange(0.001, 1, 0.001)
psi = np.arange(-20, 20, 0.01)
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,6))

ax[0].plot(theta, prior_theta(theta))
ax[0].set_xlabel("theta")
ax[0].set_ylabel("p_ J (theta)")

ax[1].plot(psi, [prior_psi(x) for x in psi])
ax[1].set_xlabel("psi")
ax[1].set_ylabel("p_ J (psi)")

plt.show()
```
